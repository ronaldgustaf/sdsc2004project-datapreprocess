```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./globalterrorismdb_0718dist.csv', encoding = 'ISO-8859-1', low_memory=False)
```

```python
# split df for before 2001 data
df_bef2001 = df[df['iyear'] < 2001]
df_bef2001
```

```python
# split df for 2001-2017 data
df_aft2001 = df[df['iyear'] >= 2001]
df_aft2001
```

```python
# get important columns
df_latlong = df.loc[:,['iyear', "country_txt", 'latitude', 'longitude']]
df_latlong.columns = ['Year', 'Country', 'Latitude', 'Longitude']
df_latlong
```

```python
# count attacks in each country
df_country_bef2001 = df_bef2001['country_txt'].value_counts().to_frame().reset_index()
df_country_bef2001.columns = ['Country', 'Terrorism Count']
df_country_bef2001
```

```python
# label each country based on the attack count -> different color in world map
conditions = [
    (df_country_bef2001['Terrorism Count'] < 100),
    (df_country_bef2001['Terrorism Count'] >= 100) & (df_country_bef2001['Terrorism Count'] < 500),
    (df_country_bef2001['Terrorism Count'] >= 500) & (df_country_bef2001['Terrorism Count'] < 1000),
    (df_country_bef2001['Terrorism Count'] >= 1000) & (df_country_bef2001['Terrorism Count'] < 2500),
    (df_country_bef2001['Terrorism Count'] >= 2500) & (df_country_bef2001['Terrorism Count'] < 5000),
    (df_country_bef2001['Terrorism Count'] >= 5000) & (df_country_bef2001['Terrorism Count'] < 7500),
    (df_country_bef2001['Terrorism Count'] >= 7500) & (df_country_bef2001['Terrorism Count'] < 10000),
    (df_country_bef2001['Terrorism Count'] >= 10000) & (df_country_bef2001['Terrorism Count'] < 12500),
    (df_country_bef2001['Terrorism Count'] >= 12500) & (df_country_bef2001['Terrorism Count'] < 15000),
    (df_country_bef2001['Terrorism Count'] >= 15000)
]

values = ['<100 Attacks', '100-499 Attacks', '500-999 Attacks', '1000-2499 Attacks', '2500-4999 Attacks', '5000-7499 Attacks','7500-9999 Attacks', '10000-12499 Attacks', '12500-14999 Attacks', '15000+ Attacks']

df_country_bef2001['Attack Count Group'] = np.select(conditions, values)

df_country_bef2001
```

```python
# merge countries with location coordinates until before 2001
df_latlong_bef2001 = df_latlong[df_latlong['Year'] < 2001]
df_finalcountry_bef2001 = pd.merge(df_latlong_bef2001, df_country_bef2001, how='left')
df_finalcountry_bef2001['Year'].astype('int32')
pd.to_datetime(df_finalcountry_bef2001['Year'], format = '%Y')
```

```python
df_finalcountry_bef2001
```

```python
# repeat same step for the year 2001-2017
df_country_aft2001 = df_aft2001['country_txt'].value_counts().to_frame().reset_index()
df_country_aft2001.columns = ['Country', 'Terrorism Count']

conditions = [
    (df_country_aft2001['Terrorism Count'] < 100),
    (df_country_aft2001['Terrorism Count'] >= 100) & (df_country_aft2001['Terrorism Count'] < 500),
    (df_country_aft2001['Terrorism Count'] >= 500) & (df_country_aft2001['Terrorism Count'] < 1000),
    (df_country_aft2001['Terrorism Count'] >= 1000) & (df_country_aft2001['Terrorism Count'] < 2500),
    (df_country_aft2001['Terrorism Count'] >= 2500) & (df_country_aft2001['Terrorism Count'] < 5000),
    (df_country_aft2001['Terrorism Count'] >= 5000) & (df_country_aft2001['Terrorism Count'] < 7500),
    (df_country_aft2001['Terrorism Count'] >= 7500) & (df_country_aft2001['Terrorism Count'] < 10000),
    (df_country_aft2001['Terrorism Count'] >= 10000) & (df_country_aft2001['Terrorism Count'] < 12500),
    (df_country_aft2001['Terrorism Count'] >= 12500) & (df_country_aft2001['Terrorism Count'] < 15000),
    (df_country_aft2001['Terrorism Count'] >= 15000)
]

values = ['<100 Attacks', '100-499 Attacks', '500-999 Attacks', '1000-2499 Attacks', '2500-4999 Attacks', '5000-7499 Attacks','7500-9999 Attacks', '10000-12499 Attacks', '12500-14999 Attacks', '15000+ Attacks']

df_country_aft2001['Attack Count Group'] = np.select(conditions, values)

df_latlong_aft2001 = df_latlong[df_latlong['Year'] >= 2001]
df_finalcountry_aft2001 = pd.merge(df_latlong_aft2001, df_country_aft2001, how='left')
df_finalcountry_aft2001['Year'].astype('int32')
pd.to_datetime(df_finalcountry_aft2001['Year'], format = '%Y')

df_finalcountry_aft2001
```

```python
# export latlong of countries to be visualized as world map in tableau
df_finalcountry_bef2001.to_excel('finalcountry_year_bef2001.xlsx')
df_finalcountry_aft2001.to_excel('finalcountry_year_aft2001.xlsx')
```

success rate

```python
# get number of total attacks for each group
df_bubble_bef = df_bef2001.loc[:,["iyear","gname","success"]]
group_bef = df_bubble_bef['gname'].value_counts()
df_group_bef = group_bef.to_frame().reset_index()
```

```python
# get number of succesful attacks for each group
df_group_bef.columns = ['Group', 'Attack Count']
df_suc_bef = df_bubble_bef.groupby(['gname'], as_index=False)['success'].sum()
df_suc_bef = df_suc_bef.sort_values(by='success', ascending=False)
df_suc_bef = df_suc_bef.reset_index(drop=True)
df_suc_bef.rename(columns = {'gname':'Group', 'success': 'Successful Attacks'}, inplace = True)
df_suc_bef
```

```python
# filter groups with at least 100 attacks
df_suc_bef = df_suc_bef[df_suc_bef['Successful Attacks'] >= 100]
df_suc_bef
```

```python
# combine columns Attack Count and Successful Attacks
df_group_bef = df_group_bef.merge(df_suc_bef,on='Group')
df_group_bef
```

```python
#calculate success rate
df_group_bef['Success Rate'] = df_group_bef['Successful Attacks']/df_group_bef['Attack Count']*100
df_group_bef
```

```python
#repeat for 2001-2017 data
df_bubble_aft = df_aft2001.loc[:,["iyear","gname","success"]]
group_aft = df_bubble_aft['gname'].value_counts()
df_group_aft = group_aft.to_frame().reset_index()
df_group_aft.columns = ['Group', 'Attack Count']
df_suc_aft = df_bubble_aft.groupby(['gname'], as_index=False)['success'].sum()
df_suc_aft = df_suc_aft.sort_values(by='success', ascending=False)
df_suc_aft = df_suc_aft.reset_index(drop=True)
df_suc_aft.rename(columns = {'gname':'Group', 'success': 'Successful Attacks'}, inplace = True)
df_suc_aft = df_suc_aft[df_suc_aft['Successful Attacks'] >= 100]
df_group_aft = df_group_aft.merge(df_suc_aft,on='Group')
df_group_aft['Success Rate'] = df_group_aft['Successful Attacks']/df_group_aft['Attack Count']*100
df_group_aft
```

```python
# export to excel to be visualized in tableau
df_group_bef.to_excel('bubble_bef.xlsx')
df_group_aft.to_excel('bubble_aft.xlsx')
```

cases and casualty per year

```python
# count total cases for each year
df_cases = df.groupby('iyear').size()
df_cases = df_cases.to_frame().reset_index()
df_cases.rename(columns = {0:'ncases'}, inplace = True)
df_cases
```

```python
# sum nkill and nwound per year
df_sum = df.loc[:,["iyear","nkill","nwound"]]
df_sum = df_sum.groupby('iyear').sum()
df_sum['nkill'] = df_sum['nkill'].astype(int)
df_sum['nwound'] = df_sum['nwound'].astype(int)
# calculate total casualty
df_sum['ncasualty'] = df_sum['nkill']+df_sum['nwound']
df_sum
```

```python
# merge dataset with ncases and nkill,nwound,ncasualty
df_final_cases = df_cases.merge(df_sum,on='iyear')
df_final_cases
```

```python
import plotly.graph_objects as go
```

```python
#visualize data using plotly graph objects
year = df_final_cases['iyear']

fig = go.Figure()

fig.add_trace(go.Bar(
    x=year,
    y=df_final_cases['ncasualty'],
    name='Casualty',
    marker_color='#f28e2b'
))

fig.add_trace(go.Scatter(
    x=year,
    y=df_final_cases['ncases'],
    name='Cases/year',
    line=dict(color='#4e79a7', width=5)
))

fig.update_layout(barmode='group', xaxis_tickangle=-90,
                title_text='<b>Number of Terrorism Attacks and Casualty from 1970 to 2017</b>',
                xaxis=dict(title="Year", type="category", linecolor="#BCCCDC"),
                yaxis=dict(linecolor="#BCCCDC"),
                width=1700, height=800, font_size=22,
                font_family="Open Sans", template="simple_white")

fig.update_yaxes(
    showgrid=True
)

fig.show()
```
